```python
# import pandas and read the CTA data in as a DataFrame
import pandas as pd
TNP_File = '../Resources/Transportation_Network_Providers-Trips.csv'

TNP_df = pd.read_csv(TNP_File)
TNP_df.count()
```

```python
# drop NA rows, drop all columns except Trip ID, timestamps, and Lat and longs
TNP_df.dropna(inplace = True)
TNP_df.drop(TNP_df.iloc[:, 3:15], inplace = True, axis = 1)
TNP_df.drop(TNP_df.columns[[1,5, 8]], inplace = True, axis = 1)
TNP_df.count()
```
